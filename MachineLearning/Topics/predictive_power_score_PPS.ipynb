# <font size=6>Predictive Power Score</font>

O Predictive Power Score (PPS) é uma métrica utilizada para medir a força preditiva entre duas variáveis, sejam elas numéricas ou categóricas. Ao contrário do coeficiente de correlação de Pearson, que é restrito a variáveis numéricas e supõe uma relação linear, o PPS é mais flexível e pode capturar relações não lineares e entre variáveis de diferentes tipos. Características do PPS:

* Escala de 0 a 1: O PPS varia de 0 a 1, onde 0 indica nenhuma capacidade preditiva e 1 indica uma capacidade preditiva perfeita.
* Assimetria: O PPS é assimétrico, ou seja, o valor do PPS de X para Y pode ser diferente do PPS de Y para X.
* Robustez: O PPS é robusto a dados faltantes e outliers, pois utiliza algoritmos de machine learning que podem lidar com essas questões.

## <font size=6>PPS Regressão

# Load Data 

```python
# load
df = pd.read_csv('data/cholesterol.csv')
```

```python
# Preview
df
```

# Processing

```python
# drop id
df.drop(columns='Id', inplace=True)

# rename columns
df.columns = [col.lower() for col in df.columns]

# replace white spaces
df.columns = [col.replace(' ', '_') for col in df.columns]

# remove accent
df.columns = [remove_acentuacao(col) for col in df.columns]

# dropna
df.dropna(inplace=True)

# One Hot encode
df = pd.get_dummies(df, columns=['grupo_sanguineo', 'fumante', 'nivel_de_atividade'])
```

```python
# View
df
```

# PPS 

```python
# Calcular PPS entre as variáveis
pps_matrix_reg = pps.matrix(df)
```

```python
# Ajustar a matrx para plot
pps_matrix_reg_pivot = pps_matrix_reg[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
```

```python
# PPS entre 2 variáveis
pps.score(df, 'peso', 'colesterol')
```

```python
# Plotar matriz de PPS
plt.figure(figsize=(10,8))
sns.heatmap(pps_matrix_reg_pivot, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Predictive Power Score\n', fontsize=16)
plt.show()
```

# Matriz de Correlação Regressão

```python
# Calculcar matriz
corr_matrix_reg = df.corr()

# Plotar matriz de correlação de Person
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix_reg, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlação\n', fontsize=16)
plt.show()
```

Utilizar PPS e Correlação de forma conjunta para verificar relações lineares e não lineares dos dados.

# Features Importance

```python
# Separar os dados
X = df.drop('colesterol', axis=1)
y = df['colesterol']
```

```python
# Definir e treinar modelo
model_lr = LinearRegression()
model_lr.fit(X,y)
```

```python
# Importância das features com base nos coeficientes do modelo
feat_importance_reg = pd.Series(model_lr.coef_, index=X.columns)
feat_importance_reg.plot(kind='barh')
plt.xlabel('Importancia')
plt.ylabel('Feature')
plt.title('Feature importance Linear Regression')
plt.show()
```

# <font size=6>PPS - Classificação

# Load Data 

```python
# load
df = pd.read_csv('data/fruit_quality.csv')
```

```python
# Preview
df
```

# Processing

```python
# drop id
df.drop(columns='A_id', inplace=True)

# rename columns
df.columns = [col.lower() for col in df.columns]

# replace white spaces
df.columns = [col.replace(' ', '_') for col in df.columns]

# remove accent
df.columns = [remove_acentuacao(col) for col in df.columns]

# dropna
df.dropna(inplace=True)

# Target to 0 1
df['quality'] = (df['quality'] == 'good').astype(int)
```

```python
# View
df
```

# PPS 

```python
# Calcular PPS entre as variáveis
pps_matrix_cls = pps.matrix(df)
```

```python
# Ajustar a matrx para plot
pps_matrix_cls_pivot = pps_matrix_cls[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
```

```python
# PPS entre 2 variáveis
pps.score(df, 'size', 'quality')
```

```python
# Plotar matriz de PPS
plt.figure(figsize=(10,8))
sns.heatmap(pps_matrix_cls_pivot, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Predictive Power Score\n', fontsize=16)
plt.show()
```

# Matriz de Correlação Classificação

```python
# Calculcar matriz
corr_matrix_cls = df.corr()

# Plotar matriz de correlação de Person
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix_cls, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlação\n', fontsize=16)
plt.show()
```

Utilizar PPS e Correlação de forma conjunta para verificar relações lineares e não lineares dos dados.

# Features Importance

```python
# Separar os dados
X = df.drop('quality', axis=1)
y = df['quality']
```

```python
# Definir e treinar modelo
model_cls = LogisticRegression()
model_cls.fit(X,y)
```

```python
# Importância das features com base nos coeficientes do modelo
feat_importance_cls = pd.Series(model_cls.coef_[0], index=X.columns)
feat_importance_cls.plot(kind='barh')
plt.xlabel('Importancia')
plt.ylabel('Feature')
plt.title('Feature importance Linear Regression')
plt.show()
```
