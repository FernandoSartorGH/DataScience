## Imports

```python
# Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Imports
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

## Data

```python
# Load data
url = "https://raw.githubusercontent.com/FernandoSartorGH/data/main/housing.csv"
df = pd.read_csv(url)

# Define X and y
X, y = df.iloc[:,0:-1], df.iloc[:,-1]
```

```python
# Separate data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

## Model

```python
# Imports
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

```python
# Define the model
model = LinearRegression()

# Fit the model
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

---

## R Squared
   * It represents the percentage of the target's (y) variance that is explained by the attributes (X).
   * Basically R2 squared calculates how must regression line is better than a mean line.
   * ###  $ \frac{explainedvariation}{totalvariation} = \frac{\sum_{i}(\hat{y_i} - {\bar{y}})^2}{\sum_{i}(y_i - \bar{y})^2} = \frac{\hat{\beta'}X'y - n\bar{Y}^2}{y'y - n\bar{Y}^2}$


#### Advantages of R Squared

* It is easily interpretable. 
* It can demonstrate how good the model is without the need for comparison with other models.

#### Disadvantages of R Squared

* The disadvantage of the R Squared is while adding new features in data the R Squared starts increasing or remains constant but it never decreases because It assumes that while adding more data variance of data increases.
* But the problem is when we add an irrelevant feature in the dataset then at that time R2 sometimes starts increasing which is incorrect

```python
# Squared R
r2 = metrics.r2_score(y_test, y_pred)
print("\n r2: %.2f%%" % (r2*100))
```

```python
# Explained variance score
# If the mean of the residuals is zero, the explained variance will be equal to r2.
ev = metrics.explained_variance_score(y_test, y_pred)
print("\n EV: %.2f%%" % (ev*100))
```

## Adjusted R squared
   * Adjusted R squared comes to control the problem with R Squared.
   * ###  $ R_{adj}^2 = 1-[(\frac{n-1}{n-k-1})*(1-R^2)]$


```python
# Adjusted Squared R
n = len(y_pred)
k = len(X_test.columns)

r2_adj = 1 - ((1-r2)*(n-1)/(n-k-1))
print("\n r2_adj: %.2f%%" % (r2_adj*100))
```

## Mean Absolute Arror
   * It represents the mean absolute prediction error. If the model were perfect, the result would be zero. However, since there is no upper limit, the metric cannot quantify how bad the model is. This does not prevent the metric from being used for a relative evaluation, that is, comparing different models. To avoid cancellations in the sum of differences, the absolute value of each difference is taken..
   * ### $ MAE =\frac{1}{n}\sum{|y - \hat{y}|} $

#### Advantages of MAE

* The MAE you get is in the same unit as the output variable.
* It is most Robust to outliers.

#### Disadvantages of MAE

* The graph of MAE is not differentiable so we have to apply various optimizers like Gradient descent which can be differentiable.

```python
# Mean absolute error
mae_train = metrics.mean_absolute_error(y_train, y_pred_train)
print("\n MAE train: %.2f" % (mae_train))

mae_test = metrics.mean_absolute_error(y_test, y_pred)
print("\n MAE test: %.2f" % (mae_test))
```

---

## Mean Squared Error

   * It represents the mean squared prediction error, which is the square of the difference between the actual and predicted values. To avoid cancellations in the sum of differences, each calculated difference is squared.
    * ### $ MSE =\frac{1}{n}\sum{(y - \hat{y})^2} $

#### Advantages of MSE

* The graph of MSE is differentiable, so you can easily use it as a loss function.

#### Disadvantages of MSE

* The value you get after calculating MSE is a squared unit of output. 
* If you have outliers in the dataset then it penalizes the outliers most and the calculated MSE is bigger. So, in short, it is not Robust to outliers which were an advantage in MAE.

```python
# Mean squared error
mse_train = metrics.mean_squared_error(y_train, y_pred_train)
print("\n MSE train: %.2f" % (mse_train))

mse_test = metrics.mean_squared_error(y_test, y_pred)
print("\n MSE test: %.2f" % (mse_test))
```

---

## Root Mean Squared Error

   * It is the square root of the mean squared error. It brings the measure back to the same unit as the data.
   * Most of the time people use RMSE as an evaluation metric and mostly when you are working with deep learning techniques the most preferred metric is RMSE.
    * ### $ MSE = \sqrt{\frac{1}{n}\sum{(y - \hat{y})^2}} $

#### Advantages of RMSE
* The output value you get is in the same unit as the required output variable which makes interpretation of loss easy.

#### Disadvantages of RMSE
* It is not that robust to outliers as compared to MAE.

```python
# Root mean squared error
root_mse_train = np.sqrt(metrics.mean_squared_error(y_train, y_pred_train))
print("\n RMSE train: %.2f" % (root_mse_train))

root_mse_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print("\n RMSE test: %.2f" % (root_mse_test))
```

## Residual plot
   * Good models will exhibit homoscedasticity. This means that the residuals are randomly distributed around zero, indicating that the variance is the same for all targets regardless of the input.

```python
# imports
import seaborn as sns
import matplotlib.pyplot as plt
```

### Residual distribuition

```python
# calc residuals
residual = y_test - y_pred

# plot
fig, ax = plt.subplots(figsize=(9,5))
plt.scatter(y=residual, x=y_pred)
plt.axhline(y = 0, color = 'k', linestyle = '-')

# Labels
plt.title('Residuals', size=18)
plt.xlabel('y_pred')
plt.ylabel('residual')
plt.grid(False)

plt.show()
```

### Residues with normal distribution

The power of Q-Q plots lies in their ability to summarize any distribution visually.

QQ plots is very useful to determine

* If two populations are of the same distribution
* If residuals follow a normal distribution. Having a normal error term is an assumption in regression and we can verify if it’s met using this.
* Skewness of distribution

In Q-Q plots, theoretical Quantile values is ploted with the sample Quantile values. Quantiles are obtained by sorting the data. It determines how many values in a distribution are above or below a certain limit. If the dataset has been generated from that distribution, we expect this chart to be close to a 45-degree line, because the sample quantiles will be similar to the theoretical quantiles. 

```python
# imports
from scipy import stats
import statsmodels.api as sm 
```

```python
# Histogram
fig, ax = plt.subplots(figsize=(9,5))

res_norm = (residual - np.mean(residual))/np.std(residual)
x = np.arange(-max(res_norm), max(res_norm), 0.01)

pd.Series(res_norm, name='Residuals').plot.hist(bins=20, ax=ax, title='Residuals')
plt.show()
```

```python
# Density
x = np.arange(-max(res_norm), max(res_norm), 0.01)

sns.kdeplot(x=res_norm, data=res_norm, common_norm=False);
plt.title("Kernel Density Function");
```

```python
# qqplot
sm.qqplot(residual, fit=True,line='45',dist=stats.norm)
plt.show()
```

## QQ Plot with Asymmetric Distribuitions

Going beyond the metrics, here are some visualizations of QQ plots with known asymmetric distributions. It's interesting to explore other distributions beyond these.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate exponential distribution data
data = np.random.exponential(scale=1, size=1000)

# Create theoretical quantiles
quantiles = np.linspace(0, 1, num=100)

# Calculate quantiles for the data
data_quantiles = np.quantile(data, quantiles)

# Create QQ plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

stats.probplot(data, dist="expon", plot=ax1)
ax1.set_title("QQ Plot - Exponential Distribution")
ax1.set_xlabel("Theoretical Quantiles")
ax1.set_ylabel("Sample Quantiles")

# Create density distribution plot
ax2.hist(data, bins=30, density=True, alpha=0.5)
x = np.linspace(0, 10, 100)
ax2.plot(x, stats.expon.pdf(x), 'r-', lw=2)
ax2.set_title("Density Distribution - Exponential Distribution")
ax2.set_xlabel("Values")
ax2.set_ylabel("Density")

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate Weibull distribution data
data = np.random.weibull(2, size=1000)

# Create theoretical quantiles
quantiles = np.linspace(0, 1, num=100)

# Calculate quantiles for the data
data_quantiles = np.quantile(data, quantiles)

# Create QQ plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

stats.probplot(data, dist="weibull_min", sparams=(2,), plot=ax1)
ax1.set_title("QQ Plot - Weibull Distribution")
ax1.set_xlabel("Theoretical Quantiles")
ax1.set_ylabel("Sample Quantiles")

# Create density distribution plot
ax2.hist(data, bins=30, density=True, alpha=0.5)
x = np.linspace(0, 3, 100)
ax2.plot(x, stats.weibull_min.pdf(x, 2), 'r-', lw=2)
ax2.set_title("Density Distribution - Weibull Distribution")
ax2.set_xlabel("Values")
ax2.set_ylabel("Density")

# Display the plots
plt.tight_layout()
plt.show()
```
