## Imports

```python
# Imports
import time
import numpy as np
from sklearn import datasets
from kmodes.kmodes import KModes
from sklearn_extra.cluster import KMedoids
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, Birch, SpectralClustering, MeanShift
```

---

## Data

```python
# Dados

# Criar datasets
n_samples = 500
seed = 30
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05, random_state=seed)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=0.05, random_state=seed)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=seed)
rng = np.random.RandomState(seed)
no_structure = (rng.rand(n_samples, 2), None)

# Dados anisotrópicos
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)

# Blobs com variância variada
varied = datasets.make_blobs(n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state)
```

```python
# Lista de datasets
datasets_list = [
    ("Noisy Circles", noisy_circles),
    ("Noisy Moons", noisy_moons),
    ("Blobs", blobs),
    ("No Structure", no_structure),
    ("Anisotropic", (X_aniso, None)),
    ("Varied", varied)
]

# Algoritmos de clustering
algorithms = [
    ("K-Means", KMeans(n_clusters=3, random_state=seed)),
    ("K-Medoid", KMedoids(n_clusters=3, random_state=seed)),
    ("K-Mode", KModes(n_clusters=3, random_state=seed)),
    ("Agglomerative", AgglomerativeClustering(n_clusters=3)),
    ("DBSCAN", DBSCAN(eps=0.3)),
    ("Birch", Birch(n_clusters=3)),
    ("Gaussian Mixture", GaussianMixture(n_components=3, random_state=seed)),
    ("Mean Shift", MeanShift()),
    ("Spectral Clustering", SpectralClustering(n_clusters=3, random_state=seed)),
    ("Agglomerative (Ward)", AgglomerativeClustering(n_clusters=3, linkage='ward'))
]
```
