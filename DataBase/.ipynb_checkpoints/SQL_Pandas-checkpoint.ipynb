```python
# Imports
import warnings
warnings.filterwarnings('ignore')

import datetime
import numpy as np
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format

# SQL
import pandasql
```

# Connection mysql

```python
import pymysql

# Create conncection
con = pymysql.connect(host = "localhost", database = "vendas_bi", user = "root", password = "dsa1234")
```

```python
# Load data
df = pd.read_sql("SELECT * FROM vendas_bi.vendas_bi", con)

con.close()

df.head()
```

## Processing

```python
# Define subset
df.drop(columns = ["row_names"], inplace = True)

# Rename columns
df.columns = ["ID_Produto", "ID_Vendedor", "Loja", "Data_Venda", "Valor_Venda"]

# dates
import datetime
df["Data_Venda"] = pd.to_datetime(df["Data_Venda"])

# ID as category
df["ID_Vendedor"] = df["ID_Vendedor"].astype('category') 
```

```python
# Separate month and year
df["Month"] = df["Data_Venda"].dt.month
df["Year"] = df["Data_Venda"].dt.year
```

## pysql

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
vend_1006 = pysqldf("select * from df where ID_Vendedor = '1006' limit 5")
vend_1006
```

```python
vend_1009 = pysqldf("select * from df where ID_Vendedor = '1009' limit 5")
vend_1009
```

```python
pysqldf("select * from vend_1006 union select * from vend_1009 order by ID_Vendedor")
```

```python
# Filter sql
pysqldf("select * from df where Loja = 'SP8821'").head()
```

```python
# Select + sum + Group_by
df.groupby(["Loja"]).sum().head()
```

```python
# Select + size + Group_by
df.groupby(["Loja"]).size().to_frame('Size').head()
```

```python
# Select + sum + Group_by
df.groupby(["Year"]).sum()
```

```python
pysqldf("SELECT Year, SUM(Valor_Venda) FROM df GROUP BY Year")
```

```python
# Select + sum + Double Group_by
df_new = df.loc[(df["ID_Vendedor"] == 1004) | (df["ID_Vendedor"] == 1005)].groupby(["Year", "ID_Vendedor"]).sum()[['Valor_Venda']]

df_new.loc[df_new['Valor_Venda']>0]
```

```python
pysqldf("SELECT Year, ID_Vendedor, SUM(Valor_Venda) FROM df WHERE ID_Vendedor = 1004 OR ID_Vendedor = 1005 GROUP BY Year, ID_Vendedor ")
```
