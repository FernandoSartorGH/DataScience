## Querying tables and executing queries

```python
# Query tables
def consultTable():

    # Create conection 
    con = mysql.connector.connect(
        host="localhost",
        user="root",
        password="dsa1234")

    # Create cursor
    cur = con.cursor()

    # Execute query
    cur.execute("SHOW DATABASES")

    for x in cur:
        print(x)
    
    # Close connection
    cur.close()
    con.close()
```

```python
# Query
consultTable()
```

```python
# Execute queries
def executeQuery(database, sql_query):
    
    con = mysql.connector.connect(
        host="localhost",
        user="root",
        password="dsa1234",
        database=database)

    cur = con.cursor()
    
    cur.execute(sql_query)
    
    # Commit
    con.commit()
        
    cur.close()
    con.close()
```

## Create base

```python
# Create base
def createBase(sql_query):
    
    con = mysql.connector.connect(
        host="localhost",
        user="root",
        password="dsa1234")

    cur = con.cursor()
    
    cur.execute(sql_query)
    
    con.commit()

    cur.close()
    con.close()
```

```python
# Create base
sql_query = "CREATE DATABASE IF NOT EXISTS mtcars"
createBase(sql_query)
```

## Create table

```python
# Create table
def createTable(database, sql_query):

    con = mysql.connector.connect(
        host="localhost",
        user="root",
        password="dsa1234",
        database=database)

    cur = con.cursor()
    
    con.commit()

    cur.execute(sql_query)
    
    cur.close()
    con.close()
```

```python
# Create table
database = 'mtcars'
sql_query = 'CREATE TABLE IF NOT EXISTS MTCars(name TEXT NOT NULL, mpg TEXT NOT NULL, cyl TEXT NOT NULL, disp TEXT NOT NULL, hp TEXT)'

# Execute
createTable(database, sql_query)
```

## Insert data

```python
# Insert data
def insertData(database, sql_query):
    
    con = mysql.connector.connect(database = database, 
                                  host = "localhost", 
                                  user = "root", 
                                  password = "dsa1234")
    
    values = []

    for i in range(0, len(df.index)):
        values.append(tuple(df.iloc[i:i+1].sum().astype(str)))
    
    cur = con.cursor()

    cur.executemany(sql_query, values)
    
    con.commit()
    
    cur.close()
    con.close()
```

```python
# Insert data
database = 'mtcars'
sql_query = """INSERT INTO mtcars (name, mpg, cyl, disp, hp) VALUES (%s, %s, %s, %s, %s)"""

insertData(database, sql_query)
```

## Query data

```python
# Query data
def queryData(database, sql_query):
    
    con = mysql.connector.connect(
        host="localhost",
        user="root",
        password="dsa1234",
        database=database)

    cur = con.cursor()
    
    cur.execute(sql_query)
               
    df_back = cur.fetchall()
    col_names = [i[0] for i in cur.description]
    df_back = pd.DataFrame(df_back, columns = [col_names])
    
    return df_back

    cur.close()
    con.close()
```

```python
# Query
database = 'mtcars'
sql_query = 'select * from MTCars'

queryData(database, sql_query)
```

## Table changes

```python
# Data changes
database = 'mtcars'
sql_query = "UPDATE MTCars SET hp = 111 WHERE name = 'Mazda RX4'"

# Execute
executeQuery(database, sql_query)

# Query
table = 'mtcars'
sql_query = 'select * from MTCars'

# Return
queryData(table, sql_query)
```

```python
# Delete data
table = 'mtcars'
sql_query = "DELETE FROM MTCars WHERE name = 'Mazda RX4'"

# Execute
executeQuery(table, sql_query)

# Query
table = 'mtcars'
sql_query = 'select * from MTCars'

# Return
queryData(table, sql_query)
```
